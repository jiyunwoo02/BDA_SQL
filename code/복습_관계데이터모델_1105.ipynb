## 1. 관계 데이터 모델

### 1) 기본 개념   
- 개념적 구조를 논리적 구조로 표현하는 논리적 데이터 모델
- 하나의 개체에 관한 데이터를 하나의 릴레리션에 저장
   
### 2) 기본 용어   
> 행(tuple), 열(속성, attribute), 도메인
   
    - 릴레이션: 하나의 개체에 관한 데이터를 2차원 테이블의 구조로 저장, 파일 관리 시스템 관점에서 파일[file]에 대응
    - 속성: 릴레이션의 열/애트리뷰트, 파일 관리 시스템에서 필드[field]에 대응
    - 투플: 릴레이션의 행, 파일 관리 시스템에서 레코드[record]에 대응
    - 도메인: 하나의 속성이 가질 수 있는 모든 값의 집합, 속성 값을 입력 및 수정할 때 적합성 판단의 기준, 일반적으로 속성의 특성을 고려한 데이터 타입으로 정의
    - 널: 속성 값을 아직 모르거나 해당되는 값이 없다 (null은 0과는 다른 내용)
    - 차수: 하나의 릴레이션에서 속성의 전체 개수 (=컬럼의 개수, 속성의 개수)
    - 카디널리티: 하나의 릴레이션에서 투플의 전체 개수

=> 고객 릴레이션의 차수=6, 카디널리티=4

### 3) 릴레이션의 구성
- 릴레이션 스키마(=릴레이션 내포): 릴레이션의 논리적 구조, 릴레이션의 이름과 릴레이션에 포함된 모든 속성의 이름으로 정의, 자주 변하지 않는 정적인 특징
- 릴레이션 인스턴스(=릴레이션 외연): 어느 한 시점에 릴레이션에 존재하는 투플들의 집합, 투플의 사입/삭제/수정이 자주 발생하는 동적인 특징

### 4) 데이터베이스의 구성
- 데이터베이스 스키마: 데베의 전체 구조, 데베 구성하는 릴레이션 스키마 모음
- 데이터베이스 인스턴싀 데베 구성하는 릴레이션 인스턴스 모음

### 5) 릴레이션의 특성
- 투플의 유일성
    - 하나의 릴레이션에는 동일한 투플이 존재할 수 X
    - 투플을 유일하게 구별 위해 선정하는 속성 or 속성 모임 값으로 판단
- 투플의 무순서
    - 하나의 릴레이션에서 투플 사이 순서 무의미
- 속성의 무순서
    - 하나의 릴레이션에서 속성 사이 순서 무의미
- 속성의 원자성
    - 속성 값으로 원자 값만 사용 가능
    - 다중 값 가질 수 없다

### 6) 키
- 릴레이션에서 투플들을 유일하게 구별하는 속성 or 속성 집합
- 기본키, 대체키, 외래키, 슈퍼키, 후보키

### 7) 키의 특성
- 유일성: 하나의 릴레이션에서 모든 투플은 서로 다른 키 값 가진다
- 최소성: 꼭 필요한 최소한의 속성들로만 키 구성

### 8) 키의 종류
- 슈퍼키
    - 유일성 만족하는 속성 or 속성 집합
    - 테이블 만들 때 제약조건 등으로 설정 가능
- 후보키
    - 유일성과 최소성 만족
    > 데이터 구조 자체가 유일값으로 잡을 수 있는 어떤 컬럼이 만약 문제 발생 => 다른 컬럼을 조합해서 식별할 수 있도록 한다
- 기본키
    - 후보키 중에서 기본적으로 사용 위해 선택
    - 유일하게 단독으로 식별 가능
- 대체키
    - 기본키로 선택되지 못한 후보키
