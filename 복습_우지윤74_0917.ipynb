2) 데이터베이스 시스템이란?
- 데이터베이스
    - 데이터를 저장해 두는 곳
- 데이터베이스 관리 시스템
    - 데이터베이스에 있는 데이터가 일관되고 무결성, 중복없이 관리 되어지게 한다
- 데이터베이스 시스템
    - 데이터베이스와 데이터베이스 관리 시스템을 이용해서 조직의 이해관계자들이 사용하게 되는 전체 시스템 [정보제공]

우리는 그동안 쌓여진 데이터만 받아보고 사용했다. 대부분 우리가 직접 데이터를 만든 적은 없다. 데이터베이스를 만든 경험이 없어서 더욱 이해가 가질 않는다.   
내가 데이터베이스를 만든다 -> "스키마"가 중요하다
- 데이터 구조와 제약 조건 정의

3) 3단계 데이터베이스 구조의 개념
> WHY 3단계?
>> 너무 복잡, 구간을 나눠서 데이터베이스의 구조를 쉽게 설명하자. 
>> 결국 데이터베이스는 기업의 모든 이해관계자가 사용하는 데 모두 같은 데이터베이스 지식을 가지고 있는 것이 아니다!
- 외부 단계
    - "내가 그 집에 들어가서 사는 사람"
    - 개별 사용자가 사용 ex) 대시보드 만들어서 보여준다
- 개념 단계
    - "만들어진 집을 관리사무소가 관리, 전기 잘 나오고 수도 나오는 것들만 중요"
    - 조직 차원에서 각 팀 차원에서 필요한 데이터 더 추린다. 개념적으로 필요한 것 세분화
- 내부 단계
    - "건물 기준, 건설업체 뼈대를 만들고 기둥 만들고,,기초 베이스 만든다"
    - 데이터베이스 설계, 우리가 필요한 데이터가 무엇이고 어디서 가져올 것인가..

위 단계로 올라가면서 추상화

4) 데이터의 독립성
데이터 독립성은 데이터베이스 관리 시스템의 중요한 장점이다.   
데이터베이스 관리 시스템이 필요한 이유이다.   
=> 하위 스키마를 변경하더라도, 상위 스키마가 영향을 받지 않는 특성
- 외부-개념 : 논리적 데이터 독립성
- 개념-내부 : 물리적 데이터 독립성

(1) 논리적 데이터 독립성 : 개념 스키마가 변경되더라도 외부 스키마가 영향을 받지 X   
(2) 물리적 데이터 독립성 : 내부 스키마가 변경되더라도 개념 스키마가 영향을 받지 X
