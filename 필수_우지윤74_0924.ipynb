## 2. 데이터베이스 사용자

1) 데이터베이스 사용자
- 데베 이용 위해 접근하는 모든 사람
- 이용 목적 따라 -> 데베 관리자/최종 사용자/응용 프로그래머
    - 데베 관리자: 데베 시스템 운영, 관리
    - 최종(일반) 사용자: 데베 접근해 데이터 조작
    - 응용 프로그래머: 데이터 언어 삽입해 응용 프로그램 작성

(1) 데베 관리자 DBA   
단순하게 쿼리만 날리는X, 데베 설계, 시스템 운영 관리 설계/구축, 운영과 유지보수 설계 구축에도 참여
- 데베 구성 요소 선정 -> 우리 조직에 필요 data?
- 데베 스키마 정의 -> 컬럼의 구성?
- 물리적 저장 구조와 접근 방법 결정 -> 논리적/물리적 데베 설계+어떤 시스템으로 접근 선택+빅쿼리,AWS,snowflake,hadoop 등 결정, 조직 차원에서는 다방면 고민(비용)
- 무결성 유지 위한 제약조건 정의 -> 데이터 중복, 잘못된 정보 안된다! => 제약조건 정확성, 유효성 유지
- 보안 및 접근 권한 정책 결정
- 백업 및 회복 기법 정의
- 시스템 데베 관리
- 시스템 성능 감시 및 성능 분석
- 데베 재구성

(2) 최종 사용자   
데이터 분석가 OR 다른 부서 실무진
- end point에서 데이터 결과 확인 OR 자기가 원하는 것만 보면 된다

(3) 응용 프로그래머   
SQL 등을 이용해 응용 프로그램을 만들고, 필요한 데이터 쉽게 찾도록 만든다
- SQL 연동 가능 부분 많다
- Python 연동, snowflake 머신러닝 연동, tableau 연동, AWS 환경 쿼리 연동

2) 데이터 언어
- DDL 데이터 정의어: 스키마 정의/수정/삭제 by쿼리
    - create,alter,drop,truncate
- DML 데이터 조작어: 데이터 삽입/수정/삭제/검색 등 처리
    - 데이터 추출 등 다양한 기본적인 쿼리 
    - select,insert,update,delete
- DCL 데이터 제어어: 내부적 규칙,기법 정의 (권한)
    - brant,revoke
- TCL: 트랜잭션 제어
    - 쿼리 작성하고 실제 프로그램이 진행되면서 일의 가장 작은 프로세스
    - save point,roll back,commit

(1) 데이터 정의어
- 스키마 구조,제약조건 변경,삭제 등을 삭제
- 내가 원하는 table을 조합해서 만들거나 OR 여러가지로 데베 구축 위한 언어
   
(2) 데이터 조작어
- 내가 원하는 데이터 추출 경우
- 인스턴스 값 출력 경우
- 정차적 데이터 조적어: 구체적으로 알려준다
- 비절차적 데이터 조작어: 큰 지시만 내린다
- > 쿼리 작성에 정해진 건X, 최소한의 쿼리문이나, 작동시간, 원리를 이해하고 최대한 효율적으로 작성할 순 있다. 정답이 있는 건 아니다. 
   
(3) 데이터 제어어
- 무결성,일관성
- 규칙과 기법 -> 다양한 사람들이 사용할 예정 약속의 개념으로 정의 -> 스키마 정의, 데이터 베이스 접근 권한 등
- 시스템 사용에 대해서 정리를 하는 것
- 데이터베이스 관리자나 권한이 많은 경우에만 부여하거나 이런 식으로 진행
- 대부분 사람들이 제어어를 접근할 순 없다. 

3) 데이터베이스 관리 시스템의 구성
- 컴파일러 Compiler : 특정 프로그래밍 언어를 다른 프로그래밍 언어로 바꿔준다


4) 데이터 모델   
현실의 문제를 -> 추상화하면서 컴퓨터가 이해할 수 있게 만든다
- "데이터 모델링은 조직의 정보 수집과 관리 시스템을 정의하는 시각적 표현 또는 청사진을 생성하는 프로세스"
- 이 청사진 또는 데이터 모델은 데이터 분석자, 과학자, 엔지니어와 같은 다양한 이해관계자들이 조직의 데이터에 대한 통일된 개념을 생성할 수 있게 도움 
- 이 모델은 해당 비즈니스가 수집하는 데이터, 서로 다른 데이터 세트 사이의 관계, 데이터를 저장하고 분석하는 데 사용되는 방식을 설명
   
> 추상화 작업
>> 현실에 있는 모든 상황 OR 존재하는 것을 모델링하는 것은 쉽지 X


*WHY 데이터 모델링 중요?*   
오늘날 조직은 다양한 소스에서 많은 양의 데이터를 수집 BUT 원시 데이터로는 충분X -> 수익성 있는 비즈니스 결정을 내리는 데 도움이 되는 실행 가능한 인사이트를 얻기 위해 데이터를 분석해야 한다! 
    
- 정확한 데이터 분석을 위해서는 효율적인 데이터 수집, 저장 및 처리가 필요, 여러 데이터베이스 기술과 데이터 처리 도구가 있으며 데이터 세트마다 효율적인 분석을 위한 도구가 다르다
- 데이터 모델링은 데이터를 이해하고 이 데이터를 저장 및 관리하기 위한 올바른 기술 선택을 할 수 있는 기회를 제공! 
- 건축가가 집을 짓기 전에 청사진을 설계하는 것과 같은 방식으로 => 비즈니스 이해관계자는 조직을 위한 데이터베이스 솔루션을 엔지니어링하기 전에! 데이터 모델을 설계
- 데이터 모델링은 다음과 같은 이점을 제공
    -  데이터베이스 소프트웨어 개발 오류 감소
    -  데이터베이스 설계 및 생성 속도와 효율성 촉진
    - 조직 전체에서 데이터 문서화 및 시스템 설계의 일관성 조성
    - 데이터 엔지니어와 비즈니스 인텔리전스 팀 간의 커뮤니케이션 촉진

(1) 개념적 모델링   
- Ex) 자동차 Cars, 고객 Customers, 대리점 정보, 대리점 직원 정보, 판매정보 테이블 => 각각이 엔터티
   
(2) 논리적 모델링   
- 스키마에 대한 정보까지 고민하면서 테이블 설계
- 엔터티들의 관계를 통해서 생성될 수 있는 테이블 고민
- Ex) 이메일 -> 소문자와 숫자만 가능, 최대 길이는 50자, @모든 소문자.com.kr
