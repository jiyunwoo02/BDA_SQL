### 1) 개체와 관계 모델
- 개체는 구별될 수 있는 모든 것을 의미한다
- 개체-관계 다이어그램 (E-R diagram)

### 2) 개체
- 형실 세계에서 조직을 운영하는 데 꼭 필요한 사람 or 사물과 같이 구별되는 것
- 저장할 만한 가치가 있는 중요한 데이터
- 개념이나 사건도 존재하는 개체가 될 수 ok
- E-R 개체를 사각형으로 표현 후 개체 이름 표기
- 속성은 타원으로 표현
     - 엔터티는 인스턴스의 집합

### 3) 속성의 분류
- 속성 == 컬럼
- 속성 값의 개수 -> 컬럼 값이 한 개야 여러 개야?
    - 단일 값 / 다중 값 속성: 인스턴스에 대해 값 1개/여러 개
- 의미 분해 가능성 -> 데이터를 더 분해해서 어떤 유의미한 데이터로 나눌 수 있는가?
    - 단순 / 복합 속성: 더 이상 분해 가능/불가능
- 기존 속성 값에서 유도 -> 파생변수
     - 유도 속성
- Unique하고 중복X 식별 가능한 "키" 속성
    - 개체 구성 속성 중에서 특별한 역할
    - 각 개체 인스턴스 식별에 사용
    - 두 개 이상으로 만들어서 사용도 ok
- Null 널: 아직 결정되지 않은, 모르는 값
    - 공백이나 0과는 다르다!

### 4) 관계
- 개체와 개체가 맺고 있는 의미 있는 연관성
- E-R 모델의 중요요소
- 개체들 사이의 대응관계, 매핑 의미
    - '아티스트는 앨범을 낸다'
    - '고객'과 '제품'에 "주문"이라는 관계
    - -> 아티스트라는 개체에 데이터가 쌓이려면 최소한 앨범은 하나가 있어야 함
- 관계를 여러 개체 사이에서 정의되는 관계 타입
- 실제 개체 인스턴스 간에 맺어진 실제 관계
- E-R 다이어그램의 마름모 표현

### 5) 관계의 유형
- 일대일
    - 남편과 아내 -> 일대일대응, 엔터티 간의 관계가 1:!
- 일대다
    - 학생과 수업 -> 학생은 여러 수업을 들을 수 있다, 관계가 1:n
- 다대다
    - 고객과 제품 -> 고객은 여러 제춤을 주문할 수 있고, 제품 입장에서는 여러 고객이 주문할 수 있다

> IF 내가 어떤 데이터 마트를 설계하는데, 실제 테이블의 속성이 일대다의 관계이다.
=> 이걸 병합하기 위해서는 어떤 식으로 접근해야 하는가?

### 6) 관계의 참여 특성
- 필수 참여
    - 모든 개체들이 관계에 반드시 참여해야 한다
    - ex) 학생들은 필수적으로 전공기초 반드시 다 들어야 한다.
- 선택 참여
    - 일부 개체만 참여해도 ok
    - 일부 인스턴스만 참여해도 ok
    - ex) 고객->주문상품 다 주문X, 원하는 상품만 주문.
- 관계의 종속성 - 구별자 or 부분키(Partial key) 강한개체와 약한개체
    - 두 개체가 관계에 대한 종속을 가진다 => "개체 B가 개체 A에 종속된다"
    - 종속 => "포함된다"
    - ex) 사원이라는 개체가 있기에 사원 가족이 만들어진다
        - 사원가족은 사원 테이블이 없다면 만들어질 수 X
        - 사원 A: 강한 개체, 사원가족 B: 약한 개체
        - 사원 --부양-- 사원가족

### 7) 논리적 데이터 모델
- 데이터베이스에 있는 데이터들 간의 관계를 표현하는 방법에 따라 다양한 모델 존재
- E-R 모델은 현실 세계 사람들의 머릿속에 그릴 수 있는 개념적인 구조 모델링
> BUT E-R 다이어그램은 표현된 개념적인 구조를 데이터베이스에 표현하는 형태를 결정하는 논리적 모델링에서는 데이터베이스 관리 시스템 종류가 중요하다.
- E-R 다이어그램 표현된 개념적 구조를 데이터베이스에 어떤 형태로 저장할 지 논리적 표현 -> 논리적 데이터 모델
